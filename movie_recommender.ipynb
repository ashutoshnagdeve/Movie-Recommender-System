# to bring genres column in proper format

```python
movies.iloc[0].genres
#.iloc[0]: This is used to retrieve the data from the first row (index 0) of the DataFrame. iloc is used for integer-based indexing.

#.genres: This seems to be accessing the "genres" column of the selected row.
#we get the op as a list of dictionaries
#desired => ["Action","Adventure"....etc]
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head()
```

# to bring  keywords column in proper format

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

#  In cast column, extracting first 3 dictionaries and in that only the name of actor

```python
def convert3(obj):
    L=[]
    counter = 0 
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
        
    
```

```python
movies["cast"] = movies["cast"].apply(convert3)
```

```python
movies.head()
```

# In crew, finding where job is director and name  

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head()
```

# converting the overview column from string to list

```python
movies["overview"]= movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

# removing spaces 

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

# creating  a column "tags" 

```python
movies["tags"] = movies["genres"] + movies["overview"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

# creating new df and keeping only tags column

```python
new_df= movies[["id","title","tags"]]
```

```python
new_df.head()
```

#  converting list into string format in " tags" column

```python
new_df["tags"] = new_df["tags"].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
new_df["tags"][0]
```

# converting into lower case

```python
new_df["tags"]= new_df["tags"].apply(lambda x: x.lower())
new_df.head()
```

# Stemming

```python
#stemmimg
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
stem("Action Adventure Fantasy ScienceFiction In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron")
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

# Text vectorization: text to vectors : bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words="english")
```

```python
cv.fit_transform(new_df["tags"]).toarray().shape
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
vectors[0]
```

```python
# to get the most frequent 5000 words
cv.get_feature_names_out()
# the problem is occurence of words like action,actions=>apply stemming 
```

# Now to calculate distance of each movie vector from  other

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
list(enumerate(similarity[0]))
```

```python
# apply sorting => reverse=True => descending
#sorted(list(enumerate(similarity[0])),reverse =True) => this will sort on basis of index, but we want on basisof second element of tuple
sorted(list(enumerate(similarity[0])),reverse =True,key= lambda x:x[1])
```

```python
# to get most similar 5 movies
sorted(list(enumerate(similarity[0])),reverse =True,key= lambda x:x[1])[1:6]
```

# Create a function which will recommend 5 similar movies 

```python
def recommend(movie):
    # to get movie index from movie name
    movie_index = new_df[new_df["title"]==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse =True,key= lambda x:x[1])[1:6]
    
    #print movies 
    for i in movies_list:
        print(new_df.iloc[i[0]].title)        
```

```python
recommend("Avatar")
```
